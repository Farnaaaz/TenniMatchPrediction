```python
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv("Clean_data.csv")

# Split the DataFrame into two parts
df_part1 = df.iloc[:7880]  # First part from row 0 to row 7880 (the ones with Null value)
df_part2 = df.iloc[7880:]  # Second part from row 7881 to the end (The one with details)

# Write the two parts to separate CSV files with the header included
df_part1.to_csv("Clean_data_part1.csv", index=False)  # Exclude index column
df_part2.to_csv("Clean_data_part2.csv", index=False)  # Exclude index column
```

```python
df1 = pd.read_csv("Clean_data_part1.csv")
df2 = pd.read_csv("Clean_data_part2.csv")
```

```python
print(df1.shape[0])
df1.head()
```

```python
print(df2.shape[0])
df2.head()
```

Formatting the names in 'Winner' and 'Loser' features

```python
# Function to extract last name and initial of first name
def process_name(name):
    # Split the name into parts
    name_parts = name.split()

    # Extract the last name
    last_name = name_parts[-1]

    # Extract the initial of the first name
    first_initial = name_parts[0][0].upper() + '.'  # Convert the initial to uppercase and add a dot

    # Concatenate the last name and initial of the first name
    processed_name = f"{last_name} {first_initial}"

    return processed_name

# Update the "Winner" and "Loser" columns
df2['Winner'] = df2['Winner'].apply(process_name)
df2['Loser'] = df2['Loser'].apply(process_name)
```

```python
# Save the updated df2 to a new CSV file
df2.to_csv("updated_df2.csv", index=False)
df2_updated = pd.read_csv("updated_df2.csv")
# Display the updated DataFrame
print(df2_updated.head())
```

Searching the data with playesr details to be added to the data with massive null values

```python
# Iterate over each row in df1
for idx, row in df1.iterrows():
    winner = row['Winner']
    loser = row['Loser']

    # Search for the winner in df2_updated
    winner_match = df2_updated[df2_updated['Winner'] == winner]
    if not winner_match.empty:
        # Copy features from df2 to df1
        df1.loc[idx, ['WRank', 'W_hand', 'W_ht', 'W_nationality', 'W_age', 'W_wt']] = \
            winner_match[['WRank', 'W_hand', 'W_ht', 'W_nationality', 'W_age', 'W_wt']].iloc[0].values

    # Search for the loser in df2_updated
    loser_match = df2_updated[df2_updated['Loser'] == loser]
    if not loser_match.empty:
        # Copy features from df2 to df1
        df1.loc[idx, ['LRank', 'L_hand', 'L_ht', 'L_nationality', 'L_age', 'L_wt']] = \
            loser_match[['LRank', 'L_hand', 'L_ht', 'L_nationality', 'L_age', 'L_wt']].iloc[0].values

# Save the updated df1 to a new CSV file
df1.to_csv("updated_df1.csv", index=False)
```

```python
df1_updated = pd.read_csv("updated_df1.csv")
df1_updated.head()
```

Merge the two files vertically

```python
# Concatenate the DataFrames vertically
players_info_added_Clean_data = pd.concat([df1_updated, df2_updated], ignore_index=True)

# Drop the index column
players_info_added_Clean_data.to_csv("players_info_added_Clean_data.csv", index=False)
```
